```python
df.bfill(inplace=True)
```

```python
df.isnull().sum()
```

```python
df_country=pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
final_df.Country.unique()
```

```python
country=final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_name
```

```python
plt.pie(country[:3],autopct="%.2f%%",labels=country_name[:3])
plt.title("Country wise Order percentage")
plt.show()
```

```python
final_df.columns
```

```python
rating=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating_count"})
```

```python
rating
```

```python
rating.head()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=rating,x="Aggregate rating",y="Rating_count", hue="Rating color",palette=['blue','red','orange','yellow','green','green'])
plt.show()
```

```python
sns.countplot(x="Rating color",data=rating ,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
